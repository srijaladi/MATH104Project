```python
all_data_5 = np.loadtxt("loss_acc_time_data_5.txt")
all_data_10 = np.loadtxt("loss_acc_time_data_10.txt")
```

```python
def train_loss_graph(data: np.ndarray):
    x_axis = np.arange(np.shape(data)[1])
    plt.plot(x_axis, data[0])
    plt.title("Training Loss Over Epochs")
    plt.show()

def train_acc_graph(data: np.ndarray):
    x_axis = np.arange(np.shape(data)[1])
    plt.plot(x_axis, data[1])
    plt.title("Training Accuracy Over Epochs")
    plt.show()
    
train_loss_graph(all_data_10)
train_acc_graph(all_data_10)
```

```python
def display_data_times(data: np.ndarray):
    x_axis = np.arange(np.shape(data)[1])
    for i in range(2,15,2):
        if i == 4:
            continue
        plt.plot(x_axis,data[i+1],label='rank:'+str((i-4)//2))
    plt.title("Time taken to test images over epochs (exclude regular convolution)")
    leg = plt.legend(loc='upper center')
    plt.show()
    
def display_data_times_reg_conv(data: np.ndarray):
    x_axis = np.arange(np.shape(data)[1])
    i = 4
    plt.plot(x_axis,data[i+1],label="Regular Convolution Numpy")
    plt.title("Time taken to test images over epochs (ONLY regular convolution)")
    leg = plt.legend(loc='upper center')
    plt.show()
    
def print_data_times_averages(data: np.ndarray):
    avg_data = np.sum(data, axis = 1)/np.shape(data)[1]
    for i in range(2,15,2):
        avg = avg_data[i+1]
        print("RANK: " + str((i-4)//2) + " Average Time: " + str(avg))

print("NOTE THAT RANK -1 CORRESPONDS TO PYTORCH DEFAULT CONVOLUTION MODEL FORWARD CALL")
print("NOTE THAT RANK 0 CORRESPONDS TO REGULAR CONVOLUTION USING NUMPY")
display_data_times(all_data_10)
display_data_times_reg_conv(all_data_10)
print_data_times_averages(all_data_10)
```

```python
def display_data_acc(data: np.ndarray):
    x_axis = np.arange(np.shape(data)[1])
    for i in range(2,15,2):
        if i == 4:
            continue
        plt.plot(x_axis,data[i],label='rank:'+str((i-4)//2))
    plt.title("Test Accuracy over epochs")
    leg = plt.legend(loc='upper center')
    plt.show()
    
def print_data_acc_averages(data: np.ndarray):
    avg_data = np.sum(data, axis = 1)/np.shape(data)[1]
    for i in range(2,15,2):
        avg = round(avg_data[i],5)
        print("RANK: " + str((i-4)//2) + " Average Test Accuracy: " + str(avg))

print("NOTE THAT RANK -1 CORRESPONDS TO PYTORCH DEFAULT CONVOLUTION MODEL FORWARD CALL")
print("NOTE THAT RANK 0 CORRESPONDS TO REGULAR CONVOLUTION USING NUMPY")
display_data_acc(all_data_10)
print_data_acc_averages(all_data_10)
```
